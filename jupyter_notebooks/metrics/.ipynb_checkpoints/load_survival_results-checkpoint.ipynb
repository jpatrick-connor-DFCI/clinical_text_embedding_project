```python
import os
import re
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from time import time
from tqdm import tqdm
from itertools import product
import warnings
warnings.filterwarnings("ignore")

import importlib.util
spec = importlib.util.spec_from_file_location('script_utils', '/PHShome/jpc91/clinical_text_project/python_scripts/script_utils.py') 
script_utils = importlib.util.module_from_spec(spec)	 
spec.loader.exec_module(script_utils)

def plot_model_performance(results_df, score_type, plot_title, figsize=(8,6), ylim=None):

    decay_param_str = [str(val) for val in results_df.sort_values(by='decay_param')['decay_param'].unique()]
    coxPH_results = results_df.loc[results_df['model'] == 'coxPH'].sort_values(by='decay_param')
    rsf_results = results_df.loc[results_df['model'] == 'rsf'].sort_values(by='decay_param')
    gbm_results = results_df.loc[results_df['model'] == 'gbm'].sort_values(by='decay_param')
    
    score_str = 'mean_' + score_type
    score_sem_str = 'sem_' + score_type
    
    model_results = {
        'CoxPH' : (coxPH_results[score_str], coxPH_results[score_sem_str]),
        'RSF' : (rsf_results[score_str], rsf_results[score_sem_str]),
        'GBM' : (gbm_results[score_str], gbm_results[score_sem_str])}

    x = np.arange(len(decay_param_str))  # the label locations
    width = 0.25  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(figsize=figsize, layout='constrained')

    for model, (score_value, score_sem) in model_results.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, score_value, width, yerr=score_sem, label=model)
        multiplier += 1

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel(score_str)
    ax.set_title(plot_title)
    ax.set_xticks(x + width, decay_param_str)
    ax.set_xlabel('decay parameter')
    ax.legend(loc='upper left', ncols=len(decay_param_str))
    if ylim:
        ymin, ymax = ylim
        ax.set_ylim(ymin, ymax)
```

```python
results_path = '/data/gusev/USERS/jpconnor/clinical_text_project/data/survival_data/results/'
```

```python
metric_cols = ['mean_c_index', 'sem_c_index', 'mean_auc(t)', 'sem_auc(t)', 'mean_ibs', 'sem_ibs']

results_cols = ['event', 'model', 'decay_param'] + metric_cols
results = []
# for event in ['death', 'irAE']:
for event in ['death']:
    event_results_path = results_path + event + '_with_decay/'
    for model in ['coxPH', 'rsf', 'gbm']:
        model_results_path = event_results_path + model + '/'
        for file in [f for f in os.listdir(model_results_path) if 'mean_w_decay' in f]:
            decay_str = '_'.join(re.split('_', file.replace('.csv', '')))
            decay_param = float(decay_str.replace('mean_w_decay_param_', ''))

            cur_file = pd.read_csv(model_results_path + file, index_col=0)[metric_cols].sort_values(by='mean_auc(t)', ascending=False)
            cur_file.insert(0, 'decay_param',  decay_param)
            cur_file.insert(0, 'model', model)
            cur_file.insert(0, 'event', event)
            
            results.append(cur_file.iloc[0].values.tolist())
            
results_df = pd.DataFrame(results, columns=results_cols)
pd.set_option('display.float_format', '{:.2g}'.format)
```
