```python
import os
import re
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from tqdm import tqdm
from time import time

import warnings
warnings.filterwarnings("ignore")

proj_path = '/data/gusev/USERS/jpconnor/clinical_text_project/'
fig_path = proj_path + 'figures/model_metrics/'
data_path = proj_path + 'data/'
surv_path = data_path + 'survival_data/'
notes_path = data_path + 'batched_datasets/VTE_data/processed_datasets/'
results_path = surv_path + 'results/'

output_path = results_path + 'phecode_model_comps_final_all_notes/'
events = os.listdir(output_path)

icd_to_phecode_map = pd.read_csv(data_path + 'code_data/icd_to_phecode_map.csv')
phecode_descr_map = dict(zip(icd_to_phecode_map['PHECODE'].astype(str), icd_to_phecode_map['PHECODE_DESCR']))
phecode_descr_map.update({'death' : 'death', 'vte' : 'vte'})
```

```python
os.listdir(results_path + 'phecode_model_comps_final/death')
```

## Stage Dataset Comparison

```python
metric_data = []
for event in events:
    event_path = output_path + event + '/'
    
    base_results = pd.read_csv(event_path + 'coxPH_base_model_metrics.csv')
    stage_results = pd.read_csv(event_path + 'coxPH_stage_model_metrics.csv')
    type_results = pd.read_csv(event_path + 'coxPH_type_model_metrics.csv')
    stage_type_results = pd.read_csv(event_path + 'coxPH_stage_type_model_metrics.csv')
    embed_type_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_type_test_metrics.csv')
    embed_stage_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_stage_test_metrics.csv')
    embed_stage_type_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_stage_type_test_metrics.csv')
    
    base_entry = base_results.loc[base_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    stage_entry = stage_results.loc[stage_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    type_entry = type_results.loc[type_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    stage_type_entry = stage_type_results.loc[stage_type_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    embed_type_entry = embed_type_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    embed_stage_entry = embed_stage_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    embed_stage_type_entry = embed_stage_type_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']

    if event in phecode_descr_map.keys():
        description = phecode_descr_map[event]
    else:
        description = event
        
    metric_data.append([event, description, base_entry, stage_entry, type_entry, stage_type_entry,
                        embed_type_entry, embed_stage_entry, embed_stage_type_entry])
    
metrics_df = pd.DataFrame(metric_data, columns=['event', 'event_descr', 'base_mean_auc(t)', 'stage_mean_auc(t)', 'type_mean_auc(t)', 'stage_type_mean_auc(t)',
                                                'text_plus_type_mean_auc(t)', 'text_plus_stage_mean_auc(t)', 'text_plus_stage_type_mean_auc(t)'])
```

```python
metrics_df = metrics_df.loc[(metrics_df['event_descr'] != 'Pancytopenia') & (metrics_df['event'] != '275.3')]
```

```python
metrics_df.loc[metrics_df['event'] == 'death', [col for col in metrics_df.columns if 'auc(t)' in col]]
```

### Performance across phecodes

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))

axes = axes.flatten()

for i, col in enumerate(['stage_mean_auc(t)', 'type_mean_auc(t)', 'stage_type_mean_auc(t)', 
                        'text_plus_stage_mean_auc(t)', 'text_plus_type_mean_auc(t)', 'text_plus_stage_type_mean_auc(t)']):
    metrics_df.sort_values(by=col, ascending=False, inplace=True)
    metrics_df['ranking'] = range(len(metrics_df))
    mean_val = metrics_df[col].mean()
    sns.scatterplot(metrics_df, x='ranking', y=col, ax=axes[i])
    axes[i].axhline(mean_val, label=f'Mean = {mean_val : 0.2f}', color='red')
    axes[i].set_title(col)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

### Comparing two models across phecodes

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

axes = axes.flatten()

for i, (x_col, y_col) in enumerate([('stage_mean_auc(t)', 'text_plus_stage_mean_auc(t)'), 
                                    ('type_mean_auc(t)', 'text_plus_type_mean_auc(t)'), 
                                    ('stage_type_mean_auc(t)', 'text_plus_stage_type_mean_auc(t)')]):
    metrics_df['avg_auc'] = metrics_df[[x_col, y_col]].mean(axis=1)
    metrics_df.sort_values(by='avg_auc', ascending=False, inplace=True)
    metrics_df['ranking'] = range(len(metrics_df))

    sns.scatterplot(metrics_df, x=x_col, y=y_col, ax=axes[i]).set(title=f'{x_col.replace('_auc(t)', '')} vs. {y_col.replace('_auc(t)', '')}')
    
    x = metrics_df[x_col]; y = metrics_df[y_col];
    
    lims= [min(min(x), min(y)) - 0.025,
           max(max(x), max(y)) + 0.025]
    axes[i].plot(lims, lims, 'r-', linewidth=2)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
metrics_df['text_vs_stage'] = metrics_df['text_plus_stage_mean_auc(t)'] - metrics_df['stage_mean_auc(t)']
metrics_df['text_vs_type'] = metrics_df['text_plus_type_mean_auc(t)'] - metrics_df['type_mean_auc(t)']
metrics_df['text_vs_stage_type'] = metrics_df['text_plus_stage_type_mean_auc(t)'] - metrics_df['stage_type_mean_auc(t)']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

axes = axes.flatten()

for i, col in enumerate(['text_vs_stage', 'text_vs_type', 'text_vs_stage_type']):
    mean_val = metrics_df[col].mean()
    sns.kdeplot(metrics_df, x=col, ax=axes[i])
    axes[i].axvline(mean_val, label=f'Mean = {mean_val : 0.2f}', color='red')
    axes[i].set_title(col)
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
# col, ax, order=[True, False], label_map=None

counts = metrics_df[col].value_counts(normalize=True).reindex([True, False], fill_value=0) * 100
labels = [str(k) for k in counts.index]

# Plot bars (x=labels ensures tick labels are in the same order as counts)
sns.barplot(x=labels, y=counts.values, ax=ax, palette="Set2")

# Annotate each bar using the bar patch positions (robust alignment)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f"{height:.1f}%", 
                xy=(p.get_x() + p.get_width() / 2, height),
                xytext=(0, 4), textcoords="offset points",
                ha="center", va="bottom", fontsize=11)

ax.set_ylim(0, 100)            # percentages
ax.set_xlabel("")
ax.set_ylabel("Percentage")
```

```python
metrics_df['better_in_text_than_stage'] = metrics_df['text_vs_stage'] > 0
metrics_df['better_in_text_than_type'] = metrics_df['text_vs_type'] > 0
metrics_df['better_in_text_than_stage_type'] = metrics_df['text_vs_stage_type'] > 0

# Plot
fig, axes = plt.subplots(1, 3, figsize=(14, 6), sharey=True)

for ax, col in zip(axes, ['better_in_text_than_stage', 'better_in_text_than_type', 'better_in_text_than_stage_type']):
    
    counts = metrics_df[col].value_counts(normalize=True).reindex([True, False], fill_value=0) * 100
    labels = [str(k) for k in counts.index]

    # Plot bars (x=labels ensures tick labels are in the same order as counts)
    sns.barplot(x=labels, y=counts.values, ax=ax, palette="Set2")

    # Annotate each bar using the bar patch positions (robust alignment)
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f"{height:.1f}%", 
                    xy=(p.get_x() + p.get_width() / 2, height),
                    xytext=(0, 4), textcoords="offset points",
                    ha="center", va="bottom", fontsize=11)

    ax.set_ylim(0, 100)            # percentages
    ax.set_xlabel("")
    ax.set_ylabel("Percentage")
    ax.set_title(col)
    
plt.tight_layout()
plt.show()
```

## Full Dataset (Comparing Text + Type vs. Just Type)

```python
output_path = results_path + 'phecode_model_comps_full_cohort_all_notes/'
events = os.listdir(output_path)

metric_data = []
for event in events:
    event_path = output_path + event + '/'
    
    base_results = pd.read_csv(event_path + 'coxPH_base_model_metrics.csv')
    embed_test_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_type_test_metrics.csv')
    
    base_entry = base_results.loc[base_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    embed_entry = embed_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']

    if event in phecode_descr_map.keys():
        description = phecode_descr_map[event]
    else:
        description = event
    
    metric_data.append([event, description, base_entry, embed_entry])
        
full_cohort_metrics_df = pd.DataFrame(metric_data, columns=['event', 'event_descr', 'type_mean_auc(t)', 'text_plus_type_mean_auc(t)']).dropna()
```

```python
full_cohort_metrics_df.loc[full_cohort_metrics_df['event_descr'] == 'death']
```

```python
# Calculate metrics
full_cohort_metrics_df['avg_auc'] = full_cohort_metrics_df[
    ['type_mean_auc(t)', 'text_plus_type_mean_auc(t)']
].mean(axis=1)
full_cohort_metrics_df = (
    full_cohort_metrics_df.sort_values(by='avg_auc', ascending=False).dropna()
)
full_cohort_metrics_df['text_plus_type_vs_type'] = (
    full_cohort_metrics_df['text_plus_type_mean_auc(t)']
    - full_cohort_metrics_df['type_mean_auc(t)']
)

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)

# --- Panel 1: Scatterplot ---
sns.scatterplot(
    data=full_cohort_metrics_df,
    x='type_mean_auc(t)',
    y='text_plus_type_mean_auc(t)',
    ax=axes[0]
)
axes[0].set_title('Type vs. Type + Text Mean AUC(t)')

x = full_cohort_metrics_df['type_mean_auc(t)']
y = full_cohort_metrics_df['text_plus_type_mean_auc(t)']

lims = [
    min(x.min(), y.min()) - 0.025,
    max(x.max(), y.max()) + 0.025
]

axes[0].plot(lims, lims, 'r-', linewidth=2)
axes[0].set_xlim(lims)
axes[0].set_ylim(lims)

# --- Panel 2: KDE plot of difference ---
mean_val = full_cohort_metrics_df['text_plus_type_vs_type'].mean()
sns.kdeplot(
    data=full_cohort_metrics_df,
    x='text_plus_type_vs_type',
    ax=axes[1],
    fill=True
)
axes[1].axvline(mean_val, label=f'Mean = {mean_val:0.2f}', color='red')
axes[1].set_title('Type vs. Type + Text')
axes[1].legend()

# --- Panel 3: Percentage barplot ---
full_cohort_metrics_df['better_in_text_than_type'] = (
    full_cohort_metrics_df['text_plus_type_vs_type'] > 0
)
counts = (
    full_cohort_metrics_df['better_in_text_than_type']
    .value_counts(normalize=True)
    .reindex([True, False], fill_value=0)
    * 100
)
labels = [str(k) for k in counts.index]

sns.barplot(x=labels, y=counts.values, ax=axes[2], palette="Set2")

# Annotate bars
for p in axes[2].patches:
    height = p.get_height()
    axes[2].annotate(
        f"{height:.1f}%",
        xy=(p.get_x() + p.get_width() / 2, height),
        xytext=(0, 4),
        textcoords="offset points",
        ha="center",
        va="bottom",
        fontsize=11
    )

axes[2].set_ylim(0, 100)
axes[2].set_xlabel("")
axes[2].set_ylabel("Percentage")
axes[2].set_title('Better in Text vs Type')

plt.tight_layout()
plt.show()
```
