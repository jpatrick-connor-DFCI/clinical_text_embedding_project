```python
import os
import re
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from tqdm import tqdm
from time import time

import warnings
warnings.filterwarnings("ignore")

proj_path = '/data/gusev/USERS/jpconnor/clinical_text_project/'
fig_path = proj_path + 'figures/model_metrics/'
data_path = proj_path + 'data/'
surv_path = data_path + 'survival_data/'
notes_path = data_path + 'batched_datasets/VTE_data/processed_datasets/'
results_path = surv_path + 'results/'

output_path = results_path + 'phecode_model_comps_v2/'
# output_path = results_path + 'phecode_model_comps/'
events = os.listdir(output_path)

icd_to_phecode_map = pd.read_csv(data_path + 'code_data/icd_to_phecode_map.csv')
phecode_descr_map = dict(zip(icd_to_phecode_map['PHECODE'].astype(str), icd_to_phecode_map['PHECODE_DESCR']))
phecode_descr_map.update({'death' : 'death', 'vte' : 'vte'})
```

## Stage Dataset Comparison

```python
metric_data = []
for event in events:
    event_path = output_path + event + '/'
    
    base_results = pd.read_csv(event_path + 'coxPH_base_model_metrics.csv', index_col=0)
    stage_results = pd.read_csv(event_path + 'coxPH_stage_model_metrics.csv', index_col=0)
    type_results = pd.read_csv(event_path + 'coxPH_type_model_metrics.csv', index_col=0)
    stage_type_results = pd.read_csv(event_path + 'coxPH_stage_type_model_metrics.csv', index_col=0)
    embed_test_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_test_metrics.csv', index_col=0)
    genomics_test_results = pd.read_csv(event_path + 'coxPH_genomics_test_metrics.csv', index_col=0)
    ehr_test_results = pd.read_csv(event_path + 'coxPH_ehr_test_metrics.csv', index_col=0)
    
    base_entry = base_results.loc[base_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    stage_entry = stage_results.loc[stage_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    type_entry = type_results.loc[type_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    stage_type_entry = stage_type_results.loc[stage_type_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    embed_entry = embed_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    genomics_entry = genomics_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    ehr_entry = ehr_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']

    if event in phecode_descr_map.keys():
        description = phecode_descr_map[event]
    else:
        description = event
    
    if ('coxPH_decayed_embeddings_plus_stage_test_metrics.csv') in os.listdir(event_path):
        embed_plus_stage_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_stage_test_metrics.csv', index_col=0)
        embed_plus_type_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_type_test_metrics.csv', index_col=0)
        embed_plus_stage_type_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_plus_stage_type_test_metrics.csv', index_col=0)

        embed_plus_stage_entry = embed_plus_stage_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
        embed_plus_type_entry = embed_plus_type_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
        embed_plus_stage_type_entry = embed_plus_stage_type_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
        embed_plus_stage_genomics_entry = embed_plus_stage_genomics_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']

        metric_data.append([event, description, base_entry, stage_entry, type_entry, stage_type_entry, genomics_entry, ehr_entry, 
                            embed_entry, embed_plus_stage_entry, embed_plus_type_entry, embed_plus_stage_type_entry])
    else:
        metric_data.append([event, description, base_entry, stage_entry, type_entry, stage_type_entry, 
                            genomics_entry, ehr_entry, embed_entry, np.nan, np.nan, np.nan])
```

```python
metrics_df = pd.DataFrame(metric_data, columns=['event', 'event_descr', 'base_mean_auc(t)', 'stage_mean_auc(t)', 'type_mean_auc(t)', 'stage-type_mean_auc(t)', 'genomics_mean_auc(t)', 
                                                'ehr_mean_auc(t)', 'text_mean_auc(t)', 'text_plus_stage_mean_auc(t)', 'text_plus_type_mean_auc(t)', 'text_plus_stage_type_mean_auc(t)'])

analyzed_phecodes = icd_to_phecode_map.loc[icd_to_phecode_map['PHECODE'].astype(str).isin(metrics_df['event'].unique())]
```

## Full Dataset (Comparing Text + Type vs. Just Type)

```python
output_path = results_path + 'phecode_model_comps_v3/'
# output_path = results_path + 'phecode_model_comps/'
events = os.listdir(output_path)

metric_data = []
for event in events:
    event_path = output_path + event + '/'
    
    base_results = pd.read_csv(event_path + 'coxPH_base_model_metrics.csv', index_col=0)
    embed_test_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_test_metrics.csv', index_col=0)
    
    base_entry = base_results.loc[base_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    embed_entry = embed_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']

    if event in phecode_descr_map.keys():
        description = phecode_descr_map[event]
    else:
        description = event
    
    metric_data.append([event, description, base_entry, embed_entry])
        
metrics_df = pd.DataFrame(metric_data, columns=['event', 'event_descr', 'type_mean_auc(t)', 'text_mean_auc(t)']).dropna()
```
