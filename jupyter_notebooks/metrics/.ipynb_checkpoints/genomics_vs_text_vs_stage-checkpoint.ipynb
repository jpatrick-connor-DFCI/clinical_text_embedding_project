```python
import os
import re
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from tqdm import tqdm
from time import time

import importlib.util
spec = importlib.util.spec_from_file_location('script_utils', '/PHShome/jpc91/clinical_text_project/python_scripts/script_utils.py') 
script_utils = importlib.util.module_from_spec(spec)	 
spec.loader.exec_module(script_utils)

import warnings
warnings.filterwarnings("ignore")

data_path = '/data/gusev/USERS/jpconnor/clinical_text_project/data/'
surv_path = data_path + 'survival_data/'
notes_path = data_path + 'batched_datasets/VTE_data/processed_datasets/'
results_path = surv_path + 'results/'

output_path = results_path + 'phecode_predictions_genomics_vs_embeddings_vs_stage-type/'
events = os.listdir(output_path)

icd_to_phecode_map = pd.read_csv(data_path + 'icd_to_phecode_map.csv')
phecode_descr_map = dict(zip(icd_to_phecode_map['PHECODE'].astype(str), icd_to_phecode_map['PHECODE_DESCR']))
phecode_descr_map.update({'death' : 'death', 'vte' : 'vte'})
```

```python
os.listdir(output_path + 'death/')
```

```python
metric_data = []
for event in events:
    event_path = output_path + event + '/'
    
    base_results = pd.read_csv(event_path + 'coxPH_base_model_metrics.csv', index_col=0)
    stage_type_results = pd.read_csv(event_path + 'coxPH_stage_type_model_metrics.csv', index_col=0)
    embed_test_results = pd.read_csv(event_path + 'coxPH_decayed_embeddings_test_metrics.csv', index_col=0)
    genomics_test_results = pd.read_csv(event_path + 'coxPH_genomics_test_metrics.csv', index_col=0)
    
    base_entry = base_results.loc[base_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    stage_type_entry = stage_type_results.loc[stage_type_results['eval_data'] == 'test_data'].iloc[0]['mean_auc(t)']
    embed_entry = embed_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    genomics_entry = genomics_test_results.sort_values(by='mean_auc(t)', ascending=False).iloc[0]['mean_auc(t)']
    
    metric_data.append([event, phecode_descr_map[event], base_entry, stage_type_entry, genomics_entry, embed_entry])
```

```python
metrics_df = pd.DataFrame(metric_data, columns=['event', 'event_descr', 'base_mean_auc(t)', 'stage-type_mean_auc(t)', 'genomics_mean_auc(t)', 'text_mean_auc(t)'])

analyzed_phecodes = icd_to_phecode_map.loc[icd_to_phecode_map['PHECODE'].astype(str).isin(metrics_df['event'].unique())]
```

```python
metrics_df['base_mean_auc(t)'] = metrics_df['base_mean_auc(t)'].apply(lambda x : round(x, 2))
metrics_df['stage-type_mean_auc(t)'] = metrics_df['stage-type_mean_auc(t)'].apply(lambda x : round(x, 2))
metrics_df['text_mean_auc(t)'] = metrics_df['text_mean_auc(t)'].apply(lambda x : round(x, 2))
metrics_df['genomics_mean_auc(t)'] = metrics_df['genomics_mean_auc(t)'].apply(lambda x : round(x, 2))

metrics_df['mean_performance'] = metrics_df[['base_mean_auc(t)', 'stage-type_mean_auc(t)', 'text_mean_auc(t)', 'genomics_mean_auc(t)']].mean(axis=1)

metrics_df = metrics_df.sort_values(by='mean_performance', ascending=False).dropna()

metrics_df['top_model'] = np.argmax(metrics_df[['base_mean_auc(t)', 'stage-type_mean_auc(t)', 'genomics_mean_auc(t)', 'text_mean_auc(t)']].values, axis=1)
metrics_df['top_model'] = metrics_df['top_model'].map({0 : 'base', 1 : 'stage-type', 2 : 'genomics', 3 : 'text'})

metrics_df['text_vs_base'] = metrics_df['text_mean_auc(t)'] - metrics_df['base_mean_auc(t)']
metrics_df['text_vs_stage-type'] = metrics_df['text_mean_auc(t)'] - metrics_df['stage-type_mean_auc(t)']
metrics_df['text_vs_genomics'] = metrics_df['text_mean_auc(t)'] - metrics_df['genomics_mean_auc(t)']
metrics_df['genomics_vs_base'] = metrics_df['genomics_mean_auc(t)'] - metrics_df['base_mean_auc(t)']
metrics_df['genomics_vs_stage-type'] = metrics_df['genomics_mean_auc(t)'] - metrics_df['stage-type_mean_auc(t)']
metrics_df['stage-type_vs_base'] = metrics_df['stage-type_mean_auc(t)'] - metrics_df['base_mean_auc(t)']
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20,10))
sns.kdeplot(metrics_df, x='text_vs_base', ax=ax[0,0])
ax[0,0].axvline(x=metrics_df['text_vs_base'].mean(), color='red', label=f'{metrics_df['text_vs_base'].mean() : 0.2f}')
ax[0,0].legend()

sns.kdeplot(metrics_df, x='stage-type_vs_base', ax=ax[0,1])
ax[0,1].axvline(x=metrics_df['stage-type_vs_base'].mean(), color='red', label=f'{metrics_df['stage-type_vs_base'].mean() : 0.2f}')
ax[0,1].legend()

sns.kdeplot(metrics_df, x='genomics_vs_base', ax=ax[0,2])
ax[0,2].axvline(x=metrics_df['genomics_vs_base'].mean(), color='red', label=f'{metrics_df['genomics_vs_base'].mean() : 0.2f}')
ax[0,2].legend()

sns.kdeplot(metrics_df, x='text_vs_stage-type', ax=ax[1,0])
ax[1,0].axvline(x=metrics_df['text_vs_stage-type'].mean(), color='red', label=f'{metrics_df['text_vs_stage-type'].mean() : 0.2f}')
ax[1,0].legend()

sns.kdeplot(metrics_df, x='text_vs_genomics', ax=ax[1,1])
ax[1,1].axvline(x=metrics_df['text_vs_genomics'].mean(), color='red', label=f'{metrics_df['text_vs_genomics'].mean() : 0.2f}')
ax[1,1].legend()

sns.kdeplot(metrics_df, x='genomics_vs_stage-type', ax=ax[1,2])
ax[1,2].axvline(x=metrics_df['genomics_vs_stage-type'].mean(), color='red', label=f'{metrics_df['genomics_vs_stage-type'].mean() : 0.2f}')
ax[1,2].legend()
```

```python
plot_order_dict = dict(zip(metrics_df['event'], range(len(metrics_df['event']))))
plot_metrics_df = metrics_df.melt(id_vars=['event', 'event_descr'], value_vars=['base_mean_auc(t)', 'stage-type_mean_auc(t)', 'genomics_mean_auc(t)', 'text_mean_auc(t)'], var_name='model', value_name='mean_auc(t)')
plot_metrics_df['model'] = plot_metrics_df['model'].apply(lambda x : x.split('_')[0])
plot_metrics_df['sorted_order'] = plot_metrics_df['event'].map(plot_order_dict)
plot_metrics_df.sort_values(by='sorted_order', inplace=True)
```

```python
text_plot_metrics_df = plot_metrics_df.loc[plot_metrics_df['model'] == 'text'].sort_values(by='mean_auc(t)', ascending=False)
text_plot_metrics_df['sorted_order'] = range(len(text_plot_metrics_df['mean_auc(t)']))

plt.figure(figsize=(12,8))
sns.lineplot(text_plot_metrics_df, x='sorted_order', y='mean_auc(t)').set(title='Text Model Performance Across Phecodes')
plt.axhline(y=0.7, color='red', label='mean_auc(t) = 0.7')
plt.legend()
```
