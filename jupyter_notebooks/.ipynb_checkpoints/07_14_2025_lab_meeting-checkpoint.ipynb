## Full cohort performance

```python
results_path = data_path + 'survival_data/results/ICD_predictions_v3/'
events = os.listdir(results_path)

models=['coxPH']
met_events = ['brainM', 'boneM', 'adrenalM', 'liverM', 'lungM', 'nodeM', 'peritonealM']
metrics_col = ['mean_c_index', 'sem_c_index', 'mean_auc(t)', 'sem_auc(t)', 'mean_ibs', 'sem_ibs']
model_performance = []
for event in events:
    for model in models:
        res_df = pd.read_csv(results_path + event + '/' + model + '_decayed_embeddings_full_data.csv', index_col=0).sort_values(by='mean_auc(t)', ascending=False)
        base_df = pd.read_csv(results_path + event + '/' + model + '_no_embeddings_full_data.csv', index_col=0).sort_values(by='mean_auc(t)', ascending=False)
        model_performance.append([event, model, 'decayed_embeddings'] + res_df[metrics_col].iloc[0].tolist())
        model_performance.append([event, model, 'no_embeddings'] + base_df[metrics_col].iloc[0].tolist())
        
full_res_df = pd.DataFrame(model_performance, columns=['event', 'model', 'data'] + metrics_col)
```

```python
import icd10
def find_icd_code(code):
    if icd10.exists(code):
        code_lookup = icd10.find(code)
        return code_lookup.description
    else:
        return code

def find_icd_block_description(code):
    if icd10.exists(code):
        try:
            return icd10.find(code).block_description
        except:
            return None
    elif code in ['brainM', 'boneM', 'adrenalM', 'liverM', 'lungM', 'nodeM', 'peritonealM']:
        return 'Metastasis'
    elif code in ['death', 'vte']:
        return code
    else:
        return None
    
full_res_df['code_description'] = full_res_df['event'].apply(find_icd_code)
full_res_df['block_description'] = full_res_df['event'].apply(find_icd_block_description)
```

```python
decayed_res = full_res_df.loc[full_res_df['data'] == 'decayed_embeddings'].sort_values(by='mean_auc(t)', ascending=False)
decayed_res['mean_auc(t)_rank'] = range(len(decayed_res))

sns.lineplot(decayed_res, x='mean_auc(t)_rank', y='mean_auc(t)').set(title='Performance Across All Events')
plt.axhline(y=0.7, color='red')
```

# Performance within cancer type

```python
intae_data_path = '/data/gusev/PROFILE/CLINICAL/robust_VTE_pred_project_2025_03_cohort/data/'
cancer_type_df = pd.read_csv(intae_data_path + 'first_treatments_dfci_w_inferred_cancers.csv')
cancer_type_df[['DFCI_MRN', 'med_genomics_merged_cancer_group']]
```

```python
os.listdir(data_path + 'survival_data/')

risk_profiles = pd.read_csv(surv_path + 'complete_risk_scores_df.csv', index_col=0)
risk_profiles['CANCER_TYPE'] = risk_profiles['DFCI_MRN'].map(dict(zip(cancer_type_df['DFCI_MRN'], cancer_type_df['med_genomics_merged_cancer_group'])))

from sksurv.metrics import concordance_index_censored

cancer_types = risk_profiles['CANCER_TYPE'].unique()

events = [re.split('_', val)[0] for val in [col for col in risk_profiles.columns if 'indicator' in col]]
metric_data=[]
for cancer_type in tqdm(cancer_types):
    cancer_risk_profile_df = risk_profiles.loc[risk_profiles['CANCER_TYPE'] == cancer_type]
    n_patients = len(cancer_risk_profile_df)
    for event in events:
        ind_col = cancer_risk_profile_df[event + '_indicator'].astype(bool)
        t_col = cancer_risk_profile_df[event + '_time']
        risk_col = cancer_risk_profile_df[event + '_risk_score']
        try:
            c_index = concordance_index_censored(event_indicator=ind_col, event_time=t_col, estimate=risk_col)[0]
        except:
            c_index=np.nan
        n_events = ind_col.sum()
        metric_data.append([cancer_type, event, n_patients, n_events, c_index])
full_metrics_df = pd.DataFrame(metric_data, columns=['cancer_type', 'event', 'n_px', 'n_events', 'c_index'])
full_metrics_df['event_descr'] = full_metrics_df['event'].apply(find_icd_code)
```

```python
filtered_metrics_df = full_metrics_df.loc[(full_metrics_df['n_px'] >= 500) & (full_metrics_df['n_events'] >=10) &
                                          (full_metrics_df['cancer_type'] != 'OTHER')].sort_values(by='c_index', ascending=False)
```

```python
base_metrics = full_res_df.loc[full_res_df['data'] == 'decayed_embeddings']
filtered_metrics_df['performance_across_types'] = filtered_metrics_df['event'].map(dict(zip(base_metrics['event'], base_metrics['mean_c_index'])))
filtered_metrics_df['change_in_c_index'] = filtered_metrics_df['c_index'] - filtered_metrics_df['performance_across_types']
```

```python
ctype_death_df = filtered_metrics_df.loc[filtered_metrics_df['event'] == 'death'].sort_values(by='c_index', ascending=False)
plt.figure(figsize=(12,8))
sns.barplot(ctype_death_df, x='cancer_type', y='c_index').set(title='Model Performance for OS within Cancer Type (c-index)')
plt.axhline(y=0.5, color='red')
```

# Performance within stage

```python
os.listdir(data_path + 'survival_data/')

risk_profiles = pd.read_csv(surv_path + 'complete_risk_scores_df.csv', index_col=0)
risk_profiles['CANCER_STAGE'] = risk_profiles['DFCI_MRN'].map(mrn_stage_dict)
risk_profiles = risk_profiles.loc[~risk_profiles['CANCER_STAGE'].isna()]

cancer_stages = risk_profiles['CANCER_STAGE'].unique()

events = [re.split('_', val)[0] for val in [col for col in risk_profiles.columns if 'indicator' in col]]
metric_data=[]
for stage in tqdm(cancer_stages):
    stage_risk_profile_df = risk_profiles.loc[risk_profiles['CANCER_STAGE'] == stage]
    n_patients = len(stage_risk_profile_df)
    for event in events:
        ind_col = stage_risk_profile_df[event + '_indicator'].astype(bool)
        t_col = stage_risk_profile_df[event + '_time']
        risk_col = stage_risk_profile_df[event + '_risk_score']
        try:
            c_index = concordance_index_censored(event_indicator=ind_col, event_time=t_col, estimate=risk_col)[0]
        except:
            c_index=np.nan
        n_events = ind_col.sum()
        metric_data.append([stage, event, n_patients, n_events, c_index])
full_metrics_df = pd.DataFrame(metric_data, columns=['cancer_stage', 'event', 'n_px', 'n_events', 'c_index'])
full_metrics_df['event_descr'] = full_metrics_df['event'].apply(find_icd_code)
```
