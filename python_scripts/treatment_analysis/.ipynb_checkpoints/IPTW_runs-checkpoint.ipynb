```python
import os
import pickle
import random
from tqdm import tqdm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sksurv.metrics import concordance_index_censored
from statsmodels.stats.multitest import multipletests
from lifelines import CoxPHFitter
from embed_surv_utils import run_grid_CoxPH_parallel, get_heldout_risk_scores_CoxPH, generate_survival_embedding_df

random.seed(42)  # set seed for reproducibility

# Paths
IO_PATH = '/data/gusev/USERS/mjsaleh/'
PROJ_PATH = '/data/gusev/USERS/jpconnor/clinical_text_project/'
FIGURE_PATH = os.path.join(PROJ_PATH, 'figures/model_metrics/')
DATA_PATH = os.path.join(PROJ_PATH, 'data/')
SURV_PATH = os.path.join(DATA_PATH, 'survival_data/')
RESULTS_PATH = os.path.join(SURV_PATH, 'results/')
NOTES_PATH = os.path.join(DATA_PATH, 'batched_datasets/VTE_data/processed_datasets/')
STAGE_PATH = '/data/gusev/PROFILE/CLINICAL/OncDRS/DERIVED_FROM_CLINICAL_TEXTS_2024_03/derived_files/cancer_stage/'
OUTPUT_PATH = os.path.join(RESULTS_PATH, 'phecode_model_comps_final')
os.makedirs(OUTPUT_PATH, exist_ok=True)

os.environ["JOBLIB_DEFAULT_WORKER_TIMEOUT"] = "600"

# Load datasets
cancer_type_df = pd.read_csv('/data/gusev/PROFILE/CLINICAL/robust_VTE_pred_project_2025_03_cohort/data/first_treatments_dfci_w_inferred_cancers.csv',
                             usecols=['DFCI_MRN', 'med_genomics_merged_cancer_group']).rename(columns={'med_genomics_merged_cancer_group' : 'CANCER_TYPE'})

tt_phecodes_df = pd.read_csv(os.path.join(SURV_PATH, 'time-to-phecode/tt_vte_plus_phecodes.csv'))
biomarker_df = pd.read_csv(os.path.join(SURV_PATH, 'IO_biomarker_discovery.csv')).drop_duplicates(subset=['DFCI_MRN'], keep='first')
treatment_df = pd.read_csv("/data/gusev/USERS/mjsaleh/profile_lines_of_rx/profile_rxlines.csv")
somatic_df = pd.read_csv(os.path.join(DATA_PATH, 'PROFILE_2024_MUTATION_CARRIERS.csv'))
```

```python
ICI_PATH = '/data/gusev/USERS/jpconnor/clinical_text_project/data/treatment_prediction/ICI_propensity/w_30_day_buffer/'
line1_preds = pd.read_csv(os.path.join(ICI_PATH, 'line_1_predictions.csv'))
```

```python
line1_mrns = line1_preds['DFCI_MRN'].unique()
line1_IO_data = biomarker_df.loc[biomarker_df['DFCI_MRN'].isin(line1_mrns)]
line1_non_IO_data = pd.get_dummies(tt_phecodes_df.loc[(tt_phecodes_df['DFCI_MRN'].isin(line1_mrns)) & 
                                       (~tt_phecodes_df['DFCI_MRN'].isin(line1_IO_data['DFCI_MRN']))]
                     .merge(somatic_df.drop(columns=['sample_id', 'cbio_sample_id', 'cbio_patient_id', 'onco_tree_code',
                                                              'briefcase', 'riker_pipeline_version', 'riker_run_version', 'CANCER_TYPE']), on='DFCI_MRN')
                     .merge(cancer_type_df, on='DFCI_MRN'), columns=['CANCER_TYPE', 'PANEL_VERSION'], drop_first=True).drop_duplicates(subset=['DFCI_MRN'], keep='first')

cols_to_include = list(set(line1_IO_data.columns) & set(line1_non_IO_data.columns))
required_cols = ['DFCI_MRN', 'tt_death', 'death']

cancer_type_cols = [col for col in cols_to_include if col.startswith('CANCER')]
panel_version_cols = [col for col in cols_to_include if col.startswith('PANEL_VERSION')]

base_vars = ['GENDER', 'AGE_AT_TREATMENTSTART'] + cancer_type_cols + panel_version_cols

biomarker_cols = [col for col in cols_to_include if col not in (required_cols + base_vars)]
line1_IO_data['PX_on_IO'] = 1
line1_non_IO_data['PX_on_IO'] = 0

interaction_IO_df = pd.concat([line1_IO_data[required_cols + base_vars + biomarker_cols + ['PX_on_IO']],
                               line1_non_IO_data[required_cols + base_vars + biomarker_cols + ['PX_on_IO']]])

interaction_IO_df['IO_prediction'] = interaction_IO_df['DFCI_MRN'].map(dict(zip(line1_preds['DFCI_MRN'], line1_preds['model_probs'])))
interaction_IO_df['IO_propensity'] = 1 / interaction_IO_df['IO_prediction']
```

```python
results = []
for test_col in tqdm(biomarker_cols):

    if interaction_IO_df[test_col].sum() > 0:
        try:
            marker_covariates = interaction_IO_df[['tt_death', 'death'] + base_vars + ['PX_on_IO', 'IO_propensity', test_col]]
            marker_covariates[test_col + '_x_IO'] = marker_covariates['PX_on_IO'] * marker_covariates[test_col]

            unweighted_cph = CoxPHFitter().fit(marker_covariates.drop(columns=['IO_propensity']), duration_col='tt_death', event_col='death')
            weighted_cph = CoxPHFitter().fit(marker_covariates, duration_col='tt_death', event_col='death', weights_col='IO_propensity')

            unweighted_summary = unweighted_cph.summary.reset_index()
            weighted_summary = weighted_cph.summary.reset_index()

            results.append([test_col] + unweighted_summary.loc[unweighted_summary['covariate'] == test_col, ['coef', 'p']].values.tolist()[0] + \
                           unweighted_summary.loc[unweighted_summary['covariate'] == test_col + '_x_IO', ['coef', 'p']].values.tolist()[0] + \
                           weighted_summary.loc[weighted_summary['covariate'] == test_col, ['coef', 'p']].values.tolist()[0] + \
                           weighted_summary.loc[weighted_summary['covariate'] == test_col + '_x_IO', ['coef', 'p']].values.tolist()[0])
        except:
            continue
            
results_df = pd.DataFrame(results, columns=['marker', 'coef', 'p', 'x_IO_coef', 'x_IO_p', 'IPTW_coef', 'IPTW_p', 'x_IO_IPTW_coef', 'x_IO_IPTW_p'])

reject, pvals_corrected, _, _ = multipletests(results_df["p"], alpha=0.05, method="fdr_bh")
interact_reject, interact_pvals_corrected, _, _ = multipletests(results_df["x_IO_p"], alpha=0.05, method="fdr_bh")

reject_IPTW, pvals_corrected_IPTW, _, _ = multipletests(results_df['IPTW_p'], alpha=0.05, method='fdr_bh')
interact_reject_IPTW, interact_pvals_corrected_IPTW, _, _ = multipletests(results_df['x_IO_IPTW_p'], alpha=0.05, method='fdr_bh')

results_df['FDR'] = pvals_corrected
results_df['x_IO_FDR'] = interact_pvals_corrected
results_df['IPTW_FDR'] = pvals_corrected_IPTW
results_df['x_IO_IPTW_FDR'] = interact_pvals_corrected_IPTW
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Plot setup ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=False)

# --- 1. Marker coefficients ---
sns.kdeplot(
    data=results_df,
    x='coef',
    label='Unweighted',
    ax=axes[0],
    fill=True,
    alpha=0.5
)
sns.kdeplot(
    data=results_df,
    x='IPTW_coef',
    label='IPTW-weighted',
    ax=axes[0],
    fill=True,
    alpha=0.5
)
axes[0].axvline(0, color='black', linestyle='--', lw=1)
axes[0].set_title('Marker Coefficient Distribution')
axes[0].set_xlabel('Coefficient')
axes[0].legend()

# --- 2. Interaction coefficients ---
sns.kdeplot(
    data=results_df,
    x='x_IO_coef',
    label='Unweighted',
    ax=axes[1],
    fill=True,
    alpha=0.5
)
sns.kdeplot(
    data=results_df,
    x='x_IO_IPTW_coef',
    label='IPTW-weighted',
    ax=axes[1],
    fill=True,
    alpha=0.5
)
axes[1].axvline(0, color='black', linestyle='--', lw=1)
axes[1].set_title('Interaction Coefficient Distribution')
axes[1].set_xlabel('Coefficient')
axes[1].legend()

plt.tight_layout()
plt.show()
```
